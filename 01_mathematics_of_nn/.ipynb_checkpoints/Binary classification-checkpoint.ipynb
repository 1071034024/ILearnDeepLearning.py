# Binary classification

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
%matplotlib inline
```

```python
N_SAMPLES = 1000
TEST_SIZE = 0.1
```

## 1. Create a dataset

```python
X, y = make_circles(n_samples=N_SAMPLES, factor=.3, noise=.10)
# y = y.reshape((y.shape[0], 1))
# X = X.T
# y = y.reshape((y.shape[0], 1)).T

# assert(X.shape == (N_SAMPLES, 2))
# assert(y.shape == (N_SAMPLES, 1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=42)
# X_train = X_train.T
# y_train = y_train.T
# X_test = X_test.T
# y_test = y_test.T

# assert(X_train.shape == (2, N_SAMPLES * (1 - TEST_SIZE)))
# assert(y_train.shape == (1, N_SAMPLES * (1 - TEST_SIZE)))
# assert(X_test.shape == (2, N_SAMPLES * TEST_SIZE))
# assert(y_test.shape == (1, N_SAMPLES * TEST_SIZE))
```

```python
# X_train = X_train.T
# y_train = y_train.reshape((y_train.shape[0], 1)).T
```

```python
print("Debug print:")
print("X_train = " + str(X_train.shape))
print("y_train = " + str(y_train.shape))
print("X_test = " + str(X_test.shape))
print("y_test = " + str(y_test.shape))
```

```python
plt.figure(figsize=(8,8))
sns.set_style("whitegrid")
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train.ravel(), s=50, cmap=plt.cm.Spectral, edgecolors='black');
```

## 2.  Build model using Keras

```python
encoding_y_train = np_utils.to_categorical(y_train)
encoding_y_test = np_utils.to_categorical(y_test)
```

```python
print("Debug print:")
print("encoding_y_train = " + str(encoding_y_train.shape))
print("encoding_y_test = " + str(encoding_y_test.shape))
```

```python
model = Sequential()
model.add(Dense(5, input_dim=2,activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=100)
```

```python
score = model.evaluate(X_test, y_test, verbose=1)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
grid = np.mgrid[-1.5:1.5:100j,-1.5:1.5:100j]
grid_2d = grid.reshape(2, -1).T
X, Y = grid
prediction_probs = model.predict_proba(grid_2d, batch_size=32, verbose=0)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
contour = ax.contourf(X, Y, prediction_probs.reshape(100, 100), cmap=cm.coolwarm)
# plt.contourf(X, Y, prediction_probs.reshape(100, 100), cmap=plt.cm.Spectral)
ax.scatter(X_test[y_test==0, 0], X_test[y_test==0, 1])
ax.scatter(X_test[y_test==1, 0], X_test[y_test==1, 1], color='r')
cbar = plt.colorbar(contour, ax=ax)
```
